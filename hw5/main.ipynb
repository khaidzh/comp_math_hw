## VI.8.9

> Функция $f(x)=\dfrac{1}{A^2-x}$ приближается на $[-4,-1]$ многочленом Лагранжа по узлам $x_0=-4, x_1=-3, x_2=-2, x_3=-1$. При каких значениях $A$ оценка погрешности в равномерной норме не превосходит $\varepsilon=10^{-5}$?

### Решение

Погрешность равна остаточному члену интерполяции в форме Лагранжа
$$f(x)-I(x)=\dfrac{f^{(4)}(\xi(x))}{4!}g(x),$$
где $\xi(x)\in[-4,-1], g(x)=(x+4)(x+3)(x+2)(x+1)$.

$$f^{(4)}(x)=-\dfrac{d^4}{dx^4}\dfrac{1}{x-A^2}=-\dfrac{4!}{(x-A^2)^5}$$

Из неопределенности зависимости $\xi(x)$ мы можем только ограничить погрешность следующим

$$\text{err}=\|f(x)-I(x)\|\le\max\limits_{\xi\in[-4,-1]}\left|\dfrac{1}{(A^2-\xi)^5}\right|\cdot\max\limits_{x\in[-4,-1]}|g(x)|$$

$$\max\limits_{\xi\in[-4,-1]}\left|\dfrac{1}{(A^2-\xi)^5}\right|=\dfrac{1}{(A^2+1)^5}$$

Найти норму $g(x)$ можно либо графически, либо аналитически
$$[g(x)]'=(x+3)(x+2)(x+1)+(x+4)(x+2)(x+1)+(x+4)(x+3)(x+1)+(x+4)(x+3)(x+2)=\\=2(2x^3+15x^2+35x+25)=4\left(x+\frac{5}{2}\right)(x^2+5x+5)=4\left(x+\frac{5}{2}\right)\left(x+\frac{5-\sqrt{5}}{2}\right)\left(x+\frac{5+\sqrt{5}}{2}\right)$$
$$g\left(-\frac{5}{2}\right)=\frac{9}{16}, g\left(-\frac{5-\sqrt{5}}{2}\right)=g\left(-\frac{5+\sqrt{5}}{2}\right)=-1\Rightarrow\max\limits_{x\in[-4,-1]}|g(x)|=1$$

$$\text{err}\le\dfrac{1}{(A^2+1)^5}\le\varepsilon\Rightarrow|A|\ge\sqrt{\varepsilon^{-1/5}-1}=\sqrt{10-1}=3$$

Ответ: $|A|\ge 3$

## Задача 5
> Реализовать составной метод трапеций численного интегрирования функций. Вычислить значение интеграла $I=\int_{\pi/2}^{3\pi/2}|\sin\sin x|dx$ с точностью $\varepsilon=10^{-4}$. Полученное значение интеграла уточнить при помощи процесса Эйткена.

Обозначим $f(x)=|\sin\sin x|,a=\pi/2,b=3\pi/2,M_2=\sup\limits_{x\in[a,b]}\left|\dfrac{d^2}{dx^2}\sin\sin x\right|\le 1$. Точное значение $I = 1.78648748195005\ldots$

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import Callable


def f(x: np.float32) -> np.float32: return np.abs(np.sin(np.sin(x)))
a, b = np.pi/2, np.pi*3/2
m2, eps = 1, 1e-4
I = 1.78648748195005

x = np.arange(a, b, np.pi/100)
plt.plot(x, np.abs(np.sin(np.sin(x))))
plt.legend(['sin(sin(x))'], loc='upper center')
plt.grid()
plt.show()
```

Составная формула трапеций на равномерной сетке $x_i=a+ih,h=(b-a)/n$ &mdash; формула Котеса
$$\int\limits_a^b f(x)dx=h\left(\dfrac{f_0+f_n}{2}+\sum_{i=1}^{n-1}f_i\right)+E_n\tag{1}$$
Положив узел в точке $c=\pi=(a+b)/2$, где она не дифференцируема, получим ошибку
$$\Delta=\dfrac{M_2}{12}(b-a)h^2=\dfrac{M_2}{12}\dfrac{(b-a)^3}{n^2}\tag{2}$$
Хотим $\Delta\le\varepsilon$, откуда
$$n\ge(b-a)^{3/2}\sqrt{\dfrac{M_2}{12\varepsilon}}\tag{3}$$

$f(x)$ строго вогнута на $[a, c]$ и $[c, b]$, тогда беря $n$ кратное 8, мы положим узел в точке $c$ для всех сеток в процессе Эйткена $\Rightarrow$ можем гарантировать существование 
$$p=\log_2\dfrac{I_{4h}-I_{2h}}{I_{2h}-I_h},\tag{4}$$
так как уменьшение шага дает прирост в формуле Котеса на строго вогнутых подынтегральных функциях $(I_{4h}<I_{2h}<I_h)$

```python
class CompositeTrapezoid:
    def __init__(self, f, a, b, m2, eps) -> None:
        self.f: Callable[[np.float32], np.float32] = f
        self._n: np.int32 = np.int32(np.ceil((b-a)**(3/2)*np.sqrt(m2/12/eps)/8)*8)
        self.a: np.float32 = a
        self.b: np.float32 = b
        self.p: np.float32 = None

    def cotes(self, n: np.int32 = 0) -> np.float32:
        if not n: n = self._n
        h: np.float32 = (self.b-self.a)/n
        s: np.float32 = (self.f(self.a)+self.f(self.b))/2
        for i in range(1, n):
            s += self.f(self.a+i*h)
        return s*h
    
    def aitken_process(self, n: np.int32 = 0) -> np.float32:
        if not n: n = self._n
        s1 = self.cotes(n)
        s2 = self.cotes(n // 2)
        s4 = self.cotes(n // 4)
        self.p = np.log2((s4-s2)/(s2-s1))
        return s1-(s2-s1)**2/(s4-2*s2+s1)
    
    @property
    def n(self) -> np.int32:
        return self._n

t = CompositeTrapezoid(f, a, b, m2, eps)

def main():
    In = t.cotes()
    Ia = t.aitken_process()

    print(f'Exact value\n\tI  = {I}...')
    print(f'Number of uniform grid nodes\n\tn  = {t.n}')
    print(f'Cotes formula\n\tIn = {In}\n\tEn = {np.abs(I-In)}')
    print(f'Aitken process\n\tp  = {t.p}\n\tI* = {Ia}\n\tE* = {np.abs(I-Ia)}')

main()
```

При $n$ кратном 8 получили $E_n\le\varepsilon,E^*/I\sim\varepsilon_\text{маш}=10^{-8}$ и $p=2$ &mdash; это совпадает с порядком аппроксимации метода трапеций для функций класса $C^2$ (это также верно для кусочно-2-гладких функций, если все точки разбиения попадают в узлы). Посмотрим, что получается для $n$ с другими остатками $8$

```python
for n in range(t.n+1, t.n+8):
    In = t.cotes(n)
    Ia = t.aitken_process(n)
    print(f'For n={n}\n\tIn = {In}\n\tEn = {np.abs(I-In)}\n\tp  = {t.p}\n\tI* = {Ia}\n\tE* = {np.abs(I-Ia)}')
```

Для нечетных $n$ неверна оценка $(2)$, хотя ошибка и уменьшилась в два раза. Это связано с тем, что наша формула при четных $n$ приближает интеграл снизу из-за вогнутости функции, при нечетном же $n$ делается небольшая добавка сверху в окрестности $c$, что приводит к уменьшению ошибки для достаточно больших $n$. Также для $n$ не кратных 8 не работает процесс Эйткена: либо $n$ не кратно 4, либо $n$ кратно 4, но значение под логарифмом в $(4)$ отрицательно. В последнем случае, если взять логарифм от абсолютного значения, то результат равен 2, что интересно, хотя значительных улучшений в уменьшении ошибке нет.

```python
n = t.n+4
s1 = t.cotes(n)
s2 = t.cotes(n // 2)
s4 = t.cotes(n // 4)
p_mod = np.log2(np.abs((s4-s2)/(s2-s1)))
print(f'n={n}\tp_mod={p_mod}')
```
